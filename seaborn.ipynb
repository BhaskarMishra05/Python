```python
import numpy as np
import seaborn as sns
import random
import pandas as pd
import matplotlib.pyplot as plt
import warnings as warnings
warnings.filterwarnings("ignore")
```

```python
x=np.random.ranf(27)
y=np.random.ranf(27)
df=pd.DataFrame({"x":x,"y":y})
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x="x",y="y",data=df)
plt.xticks([])
plt.yticks([])
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(x="x",y="y",data=df)
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.boxplot(x="x",y="y",data=df,color="blue")
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.histplot(df["x"],bins=10)
```

```python
sns.violinplot(x="x",y="y",data=df)
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
sns.kdeplot(df["x"],shade=True,color="red")
plt.show()
```

```python
sns.swarmplot(x="x",y="y",data=df,alpha=1)
plt.xticks([])
plt.yticks([])
plt.show()
```

## Matrix and heatmap

```python
sns.heatmap(df.corr(),annot=True,cmap="coolwarm")
```

```python
plt.figure(figsize=(10,5))
sns.clustermap(df.corr(),annot=True)
```

## Pair and Joint Plots


```python
sns.pairplot(df)
```

```python
sns.jointplot(x="x",y="y",data=df)
```

## Regression plots


```python
sns.regplot(x="x",y="y",data=df,color="red")
plt.show()
```

```python
sns.lmplot(x="x",y="y",data=df)
```

## Categorial Plots

```python
sns.catplot(x="x",y="y",data=df,kind="box")
plt.xticks([])
plt.yticks([])
```
